# Police Calls Average Duration

Data obtained from Open Baton Rouge for Crime Incidents in Baton Rouge

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Opening the Police Incidents data set to obtain the duration of police incidents
police = pd.read_csv("Baton_Rouge_Crime_Incidents.csv")
police.head()
```

```python
police.columns
```

```python
#Filtering only useful dates and times for Police Incidents
columns = ['CHARGE DATE', 'REPORT DATE']
police[columns] = police[columns].apply(pd.to_datetime)
police
```

```python
#Obtaining the total duration of Police Incidents by substracting the report date from the charge date
police['DURATION'] = police['REPORT DATE'] - police['CHARGE DATE']
police.head()
```

```python
police['DURATION'].min()
```

```python
police['DURATION'].max()
```

```python
#Obtaining minimum, maximum, and mean values for Police Incidents
police['DURATION'].describe()
```

```python
#Converting the total duration to minutes
police['DURATION'] = police['DURATION'].dt.total_seconds() / 60
police.head()
```

```python
police['DURATION'].min()
```

```python
police['DURATION'].max()
```

```python
#Filtering the duration of Police Incidents, considering only incidents that lasted more than 10 minutes and less than 24 hours
police = police[(police['DURATION'] > 10) & (police['DURATION'] < 1440) ]
police
```

```python
#Obtaining the mean duration of Police Incidents
police['DURATION'].describe()
```
