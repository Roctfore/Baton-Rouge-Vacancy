# Fire Calls Average Duration

Data obtained from Open Baton Rouge for Fire Incidents in Baton Rouge

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Opening the Fire Incidents data set to obtain the duration of fire incidents
fire = pd.read_csv("Baton_Rouge_Fire_Incidents.csv")
fire.head()
```

```python
fire.columns
```

```python
#Filtering only useful dispatch dates and times for Fire Incidents
columns = ['DISPATCH DATE', 'DISPATCH TIME', 'CONTROL DATE', 'CONTROL TIME', 'CLEAR DATE', 'CLEAR TIME']
fire[columns] = fire[columns].apply(pd.to_datetime)
fire
```

```python
#Obtaining the total duration of Fire Incidents by substracting the dispatch time from the clear time
fire['DURATION'] = fire['CLEAR TIME'] - fire['DISPATCH TIME']
fire.head()
```

```python
#Converting the total duration to minutes
fire['DURATION'] = fire['DURATION'].dt.total_seconds() / 60
fire.head()
```

```python
fire['DURATION'].min()
```

```python
fire['DURATION'].max()
```

```python
#Filtering the duration of fire incidents, considering only incidents that lasted more than 10 minutes
fire = fire[fire['DURATION'] > 10]
fire
```

```python
#Obtaining the mean duration of Fire Incidents
fire['DURATION'].describe()
```
